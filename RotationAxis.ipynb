```python
def rotationx(psi:float,x:list)-> list:
    rotation_matrix=[[1,0,0],[0,math.cos(psi),-math.sin(psi)],[0,math.sin(psi),math.cos(psi)]]
    return np.matmul(rotation_matrix,x)
def rotationy(psi:float,x:list)-> list:
    rotation_matrix=[[math.cos(psi),0,math.sin(psi)],[0,1,0],[-math.sin(psi),0,math.cos(psi)]]
    return np.matmul(rotation_matrix,x)
def rotationz(psi:float,x:list)-> list:
    rotation_matrix=[[math.cos(psi),-math.sin(psi),0],[math.sin(psi),math.cos(psi),0],[0,0,1]]
    return np.matmul(rotation_matrix,x)
def rotation_matrixzyx(z_angle:float,y_angle:float,x_angle:float):
    z_matrix=zaxis_matrixrotation(z_angle)
    y_matrix=yaxis_matrixrotation(y_angle)
    x_matrix=xaxis_matrixrotation(x_angle)
    return np.matmul(z_matrix,np.matmul(y_matrix,x_matrix))
def rotation_zyx(rotation_matix, point):
    return np.matmul(rotation_matix,point)
def yaxis_matrixrotation(angle:float):
    RotationMatrix= np.array([[math.cos(angle),0,math.sin(angle)],[0,1,0],[-math.sin(angle),0,math.cos(angle)]])
    return RotationMatrix
def zaxis_matrixrotation(angle:float):
    RotationMatrix= np.array([[math.cos(angle),-math.sin(angle),0],[math.sin(angle),math.cos(angle),0],[0,0,1]])
    return RotationMatrix
def xaxis_matrixrotation(angle:float):
    RotationMatrix= np.array([[1,0,0],[0,math.cos(angle),-math.sin(angle)],[0,math.sin(angle),math.cos(angle)]])
    return RotationMatrix
```

```python
def spin3d(z_angle:float,y_angle:float,x_angle:float,time:int):
    angle=[z_angle,y_angle,x_angle]
    #main_points
    base_point=[0,0,0]
    end_point=[0,0,3]
    h1=1
    r1=1
    h2=1 
    r2=1/2
    round_1 = np.linspace(0, 2*math.pi, 200)
    first_cycle:list  = []
    second_cycle:list = []
    color_change_frequency=10
    k=0
    for i in round_1:
        first_cycle.append([r1*math.cos(i),r1*math.sin(i),h1])
        second_cycle.append([r2*math.cos(i),r2*math.sin(i),h2])
    #####
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')

    z_axis = np.linspace(0, 4, 100)
    xy_axis = np.linspace(-4, 4, 100)
    z_0=np.linspace(0, 0, 100)
    end_point_plot = np.linspace(10, 10, 200)
    ax.plot(xy_axis, z_0, z_0,color = 'black', linewidth=1)
    ax.plot(z_0, z_0,z_axis,color = 'black', linewidth=1)
    ax.plot(z_0, xy_axis, z_0,color = 'black', linewidth=1)
    #ax.view_init(5,30,0)

    ###
    first_cycle_rotation:list = [] 
    second_cycle_rotation:list = [] 
    #rotation points
    rot_matrix = rotation_matrixzyx(z_angle,y_angle,x_angle)

    i=0
    while i < len(first_cycle):
        first_cycle_rotation.append(rotation_zyx(rot_matrix,first_cycle[i]))
        second_cycle_rotation.append(rotation_zyx(rot_matrix,second_cycle[i]))
        i=i+1
    first_cycle=first_cycle_rotation
    second_cycle=second_cycle_rotation
    base_point = rotation_zyx(rot_matrix,base_point)
    end_point = rotation_zyx(rot_matrix,end_point)
    i=0
    lista_test1=[]
    while i < len(first_cycle)-1:
        #1level
        xr1=[base_point[0],first_cycle[i][0],first_cycle[i+1][0]]
        yr1=[base_point[1],first_cycle[i][1],first_cycle[i+1][1]]
        zr1=[base_point[2],first_cycle[i+1][2],first_cycle[i+1][2]]
        verts1 = [list(zip(xr1,yr1,zr1))]
        #2level
        xr2=[first_cycle[i][0],second_cycle[i][0],second_cycle[i+1][0],first_cycle[i+1][0]]
        yr2=[first_cycle[i][1],second_cycle[i][1],second_cycle[i+1][1],first_cycle[i+1][1]]
        zr2=[first_cycle[i][2],second_cycle[i+1][2],second_cycle[i+1][2],first_cycle[i+1][2]]
        verts2 = [list(zip(xr2,yr2,zr2))]
        #3level
        xr3=[end_point[0],second_cycle[i][0],second_cycle[i+1][0]]
        yr3=[end_point[1],second_cycle[i][1],second_cycle[i+1][1]]
        zr3=[end_point[2],second_cycle[i+1][2],second_cycle[i+1][2]]
        verts3 = [list(zip(xr3,yr3,zr3))]
        ##
        if_list = [i % color_change_frequency, round(i/color_change_frequency)]
        lista_test1.append(if_list)
        if if_list[1] % 2 == 0:
            ax.add_collection3d(Poly3DCollection(verts1,facecolors='orange', linewidths=1))
            ax.add_collection3d(Poly3DCollection(verts2,facecolors='orange', linewidths=1))
            ax.add_collection3d(Poly3DCollection(verts3,facecolors='orange', linewidths=1))
        if if_list[1] % 2 == 1:
            ax.add_collection3d(Poly3DCollection(verts1,facecolors='skyblue', linewidths=1))
            ax.add_collection3d(Poly3DCollection(verts2,facecolors='skyblue', linewidths=1))
            ax.add_collection3d(Poly3DCollection(verts3,facecolors='skyblue', linewidths=1))
        i=i+1

    eulerz = str(round(z_angle,4))
    eulery = str(round(y_angle,4))
    eulerx = str(round(x_angle,4))
    plt.title('Kąt obrotu (' + eulerz + ', '+ eulery + ', '+ eulerx+')')
    
    plt.savefig(f'./spin_img/spinimg_{time}.png',
                dpi=400, 
                transparent = False,  
                facecolor = 'white')
 
    plt.close()
```
