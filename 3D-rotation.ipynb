##### Obrot macierz 

```python
def xaxis_matrixrotation(angle:float):
    RotationMatrix= np.array([[1,0,0],[0,math.cos(angle),-math.sin(angle)],[0,math.sin(angle),math.cos(angle)]])
    return RotationMatrix
```

```python
def yaxis_matrixrotation(angle:float):
    RotationMatrix= np.array([[math.cos(angle),0,math.sin(angle)],[0,1,0],[-math.sin(angle),0,math.cos(angle)]])
    return RotationMatrix
def zaxis_matrixrotation(angle:float):
    RotationMatrix= np.array([[math.cos(angle),-math.sin(angle),0],[math.sin(angle),math.cos(angle),0],[0,0,1]])
    return RotationMatrix
def xaxis_matrixrotation(angle:float):
    RotationMatrix= np.array([[1,0,0],[0,math.cos(angle),-math.sin(angle)],[0,math.sin(angle),math.cos(angle)]])
    return RotationMatrix
```

```python
def zaxis_matrixrotation(angle:float):
    RotationMatrix= np.array([[math.cos(angle),-math.sin(angle),0],[math.sin(angle),math.cos(angle),0],[0,0,1]])
    return RotationMatrix
```

```python
def rotation_matrixzyx(z_angle:float,y_angle:float,x_angle:float):
    z_matrix=zaxis_matrixrotation(z_angle)
    y_matrix=yaxis_matrixrotation(y_angle)
    x_matrix=xaxis_matrixrotation(x_angle)
    return np.matmul(z_matrix,np.matmul(y_matrix,x_matrix))
```

```python
def rotation_zyx(rotation_matix, point):
    return np.matmul(rotation_matix,point)
```

```python
def pyramid(z_angle:float,y_angle:float,x_angle:float,time:int):
    angle=[z_angle,y_angle,x_angle]
    #main_points
    point_1 = [-2,-2,0] 
    point_2 = [-2,2,0]
    point_3 = [2,-2,0]
    point_4 = [2,2,0]
    point_5 = [0,0,5]
    #####
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    z = np.linspace(-6, 6, 100)
    z_0=np.linspace(0, 0, 100)
    ax.plot(z_0, z_0,z)
    ax.plot(z, z_0, z_0)
    ax.plot(z_0, z, z_0)
    ###
    points:list= [point_1,point_2,point_3,point_4,point_5]
    rot_points:list = [] 
    #rotation points
    rot_matrix = rotation_matrixzyx(z_angle,y_angle,x_angle)

    for i in points:
        rot_points.append(rotation_zyx(rot_matrix,i))
    for i in rot_points:
        ax.scatter3D(i[0], i[1], i[2], color='black')
    for i in points:
        ax.scatter3D(i[0], i[1], i[2], color='black')
    

    ###
    point_1 = rot_points[0]
    point_2 = rot_points[1]
    point_3 = rot_points[2]
    point_4 = rot_points[3]
    point_5 = rot_points[4]
    ###
    ax.plot3D([point_1[0],point_2[0]], [point_1[1],point_2[1]], [point_1[2],point_2[2]], 'black')
    ax.plot3D([point_1[0],point_3[0]], [point_1[1],point_3[1]], [point_1[2],point_3[2]], 'black')
    ax.plot3D([point_1[0],point_5[0]], [point_1[1],point_5[1]], [point_1[2],point_5[2]], 'black')
    ###
    ax.plot3D([point_4[0],point_2[0]], [point_4[1],point_2[1]], [point_4[2],point_2[2]], 'black')
    ax.plot3D([point_4[0],point_3[0]], [point_4[1],point_3[1]], [point_4[2],point_3[2]], 'black')
    ax.plot3D([point_4[0],point_5[0]], [point_4[1],point_5[1]], [point_4[2],point_5[2]], 'black')
    ##
    ax.plot3D([point_3[0],point_5[0]], [point_3[1],point_5[1]], [point_3[2],point_5[2]], 'black')
    ax.plot3D([point_2[0],point_5[0]], [point_2[1],point_5[1]], [point_2[2],point_5[2]], 'black')


    point_1 = [-2,-2,0] 
    point_2 = [-2,2,0]
    point_3 = [2,-2,0]
    point_4 = [2,2,0]
    point_5 = [0,0,5]
    ax.plot3D([point_1[0],point_2[0]], [point_1[1],point_2[1]], [point_1[2],point_2[2]], 'red',alpha=0.7)
    ax.plot3D([point_1[0],point_3[0]], [point_1[1],point_3[1]], [point_1[2],point_3[2]], 'red',alpha=0.7)
    ax.plot3D([point_1[0],point_5[0]], [point_1[1],point_5[1]], [point_1[2],point_5[2]], 'red',alpha=0.7)
    ###
    ax.plot3D([point_4[0],point_2[0]], [point_4[1],point_2[1]], [point_4[2],point_2[2]], 'red',alpha=0.7)
    ax.plot3D([point_4[0],point_3[0]], [point_4[1],point_3[1]], [point_4[2],point_3[2]], 'red',alpha=0.7)
    ax.plot3D([point_4[0],point_5[0]], [point_4[1],point_5[1]], [point_4[2],point_5[2]], 'red',alpha=0.7)
    ##
    ax.plot3D([point_3[0],point_5[0]], [point_3[1],point_5[1]], [point_3[2],point_5[2]], 'red',alpha=0.7)
    ax.plot3D([point_2[0],point_5[0]], [point_2[1],point_5[1]], [point_2[2],point_5[2]], 'red',alpha=0.7)
    eulerz = str(round(z_angle,4))
    eulery = str(round(y_angle,4))
    eulerx = str(round(x_angle,4))
    plt.title('Kat obrotu (' + eulerz + ', '+ eulery + ', '+ eulerx+')')
    
    plt.savefig(f'./img3D/img_{time}zyx.png', 
                transparent = False,  
                facecolor = 'white')
 
    plt.close()
```
