```python
def f(x, y):
    return np.sqrt(x ** 2 + y ** 2)

fig = plt.figure()
ax = plt.axes(projection='3d')


u, v = np.mgrid[0:2*np.pi:100j, 0:np.pi:80j]
x = 1/2*np.cos(u)*np.sin(v)
y = 1/2*np.sin(u)*np.sin(v)
z = f(x, y)

ax.plot_surface(x, y, z, cmap=cm.coolwarm)

ax.plot_surface(x, y, 1-z, cmap=cm.coolwarm)


z = np.linspace(-2, 2, 100)
z_0=np.linspace(0, 0, 100)
ax.plot(z_0, z_0,z)
ax.set_ylabel("y")
ax.set_zlabel("z")


plt.show()
plt.show()
plt.show()
```
