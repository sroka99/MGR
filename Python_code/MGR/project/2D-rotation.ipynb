```python
import math
import numpy as np
import matplotlib.pyplot as plt
import imageio
```

```python
def rotation(point, angle:float):
    RotationMatrix= np.array([[math.cos(angle),-math.sin(angle)],[math.sin(angle),math.cos(angle)]])
    return np.matmul(RotationMatrix,point)
```

```python
def reflection(point, angle:float):
    RotationMatrix= np.array([[math.cos(angle),-math.sin(angle)],[math.sin(angle),-math.cos(angle)]])
    return np.matmul(RotationMatrix,point)
```

```python
def PlotT():
    fig = plt.figure()
    axes=fig.add_axes([0,0,1,1]) 
    #
    plt.axvline(x=0, c="black", label="x=0")
    plt.axhline(y=0, c="black", label="y=0")
    axes.set_aspect('equal', adjustable='box')
    #
    plt.plot(-13, -13, marker="o", markersize=5, color= 'black', alpha = 0) 
    plt.plot(13, 13, marker="o", markersize=5, color= 'black', alpha = 0) 

    points:list = []
    x1 = np.array([10,1])
    x2 = np.array([11,1])
    x3 = np.array([11,6])
    x4 = np.array([13,6])
    x5 = np.array([13,7])
    x6 = np.array([8,7])
    x7 = np.array([8,6])
    x8 = np.array([10,6])
    #
    for i in x1,x2,x3,x4,x5,x6,x7,x8:
        points.append(i)
    #linie laczace
    t:int = 0
    while t< len(points)-1:
        first_point:list = points[0]
        point_start:list = points[t]
        point_end:list = points[t+1]
        plt.plot([point_start[0],point_end[0]],[point_start[1],point_end[1]],color= 'red')
        t=t+1
        if t == len(points)-1:
            plt.plot([first_point[0],point_end[0]],[first_point[1],point_end[1]],color= 'red')
    #punkty
    for i in points:
        plt.plot(i[0], i[1], marker="o", markersize=5, color= 'red') 
```

#### Obrot postac zespolona obrotu: $e^{i\theta}$

```python
def multiComplex(a1:list,a2:list) -> list:
    x1:float=a1[0]*a2[0] - a1[1]*a2[1]
    x2:float=a1[0]*a2[1] + a1[1]*a2[0]
    return [x1,x2]

def rotation_element(theta:float)-> list:
    return [math.cos(theta),math.sin(theta)] 
```

```python
def complex_rotation(element:list,angle:float)-> list:
    return multiComplex(element,rotation_element(angle))
```
