```python
import random    
import math
import matplotlib.pyplot as plt
```

```python
def dis_e(p1,p2):
    return ((p1[0]-p2[0])**2 + (p1[1]-p2[1])**2)**(1/2)
```

```python
def mid_point(start:list,end:list)->list:
    x_mid=(start[0] + end[0])/2
    y_mid=(start[1] + end[1])/2
    return [x_mid,y_mid]
```

```python
def mid() ->list:
    x:float = random.random()*10
    y:float = random.random()*10
    return [x,y]
```

```python
def angle():
    ###dlugosc igly
    l=1
    ###DEFINICJE 
    lista_xy1=[]
    lista_xy2=[]
    lista_xyfinal=[]
    mid_point= mid()
    x_1 = random.random()*10
    y_1 = random.random()*10
    x_2 = random.random()*10
    y_2 = random.random()*10
    ### dodawanie elementow do 1 listy
    lista_xy1.append(x_1) 
    lista_xy1.append(y_1)
    lista_xy2.append(x_2) 
    lista_xy2.append(y_2)
    #### obliczenie elementow 2 z talesa  
    deltax=x_1-x_2
    deltay=abs(y_1-y_2)
    dis=dis_e(lista_xy1,lista_xy2)
    x_new =  l*(lista_xy1[0]-lista_xy2[0])/dis
    y_new =  l*(lista_xy1[1]-lista_xy2[1])/dis
    lista_xyfinal.append(x_1+x_new)
    lista_xyfinal.append(y_1+y_new)
    final_list = [lista_xy1,lista_xyfinal]
    return final_list
```

```python
def new_place()->list:
    a,b=angle()
    #print(a,b)
    a_mid=mid_point(a,b)
    new_mid_point = mid() 
    #print(new_mid_point)
    delta_x = new_mid_point[0]
    delta_y = new_mid_point[1]
    a_x=a[0]-delta_x
    b_x=b[0]- delta_x
    a_y=a[1]-delta_y
    b_y=b[1]- delta_y
    return [[a_x,a_y],[b_x,b_y]]
```

```python
def test(lista1):
    t=0
    k=-11
    delta =1
    listx1y1=lista1[0]
    listx2y2=lista1[1]
    while k<=12:
        if (listx1y1[0] < k and listx2y2[0]>=k) or (listx1y1[0] > k and listx2y2[0]<=k):# or abs(listx2y2[0]-k)<0.01 or abs(listx2y2[0]-k)<0.01:
            #print(listx1y1[0],listx2y2[0],k)
            t=1
            break
        k=k+delta
    return t
```

```python
def esty_prawd(n):
    lista_prawdo=[]
    i=0
    while i < n:
        p1p2 = angle()
        p1=p1p2[0]
        p2=p1p2[1]
        t = test(p1p2)
        lista_prawdo.append(t)
        i=i+1
    return sum(lista_prawdo)/i
```

```python
def fun(i):
    esty = 2/(esty_prawd(i))
    return [abs(esty - (math.pi)),esty]

def average(lst):
    return sum(lst) / len(lst)
```
